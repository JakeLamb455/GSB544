```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.basketball-reference.com/friv/birthdays.fcgi?mbonth=2&day=22")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
import time
players = []
month = 1
while month <= 12:
    month_str = str(month)
    day = 1
    max_day = 31
    if month == 2:
        max_day = 29
    elif month in [4, 6, 9, 11]:
        max_day = 30

    while day <= max_day:
        day_str = str(day)
        url = "https://www.basketball-reference.com/friv/birthdays.fcgi?month=" + month_str + "&day=" + day_str
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            rows = soup.find_all('tr')
            for row in rows:
                cells = row.find_all('td')
                if len(cells) > 1:
                    names = cells[0].text
                    birth_year = cells[1].text
                    players.append({"Player Name": names, 'Month': month, 'Day': day, "Year of Birth": birth_year})
        else:
            print(f"{month}/{day}")
        day += 1
        time.sleep(3)
    month += 1
```

```python
All_players = pd.DataFrame(players)
```

```python
All_players[['Name', 'Other']] = All_players['Player Name'].str.split('*', n=1, expand = True)
```

```python
All_players = All_players[['Name', 'Month', 'Day', 'Year of Birth']]
```

```python
All_players['Date'] = All_players['Month'].astype(str) + '/' + All_players['Day'].astype(str)
```

```python
All_players = All_players[['Name', 'Date', 'Year of Birth']]
```

```python
All_players.to_csv('Player_Birthdays.csv')
```
