```python
echo = True
import pandas as pd
import numpy as np
from plotnine import *
```

```python
avo = pd.read_csv("avocado-updated-2020.csv").rename(columns = {'4046':'small', '4225': 'large', '4770': 'xlarge'})
avo.head()
```

Question 1.

Description of columns: The date column is when this observation of Avocados was observed. The AveragePrice is the average price of a single avocado. The type is the type of Avocado (conventional or organic). The year is when the observation occured. The region is the city, state, or region of the observation. Total volume is the number of avos sold in each observation. The numbered variables are the total number of Avos sold with that PLU. The bag variables are how many of each type of bags are sold.

Question 2.

```python
major_regions = ['Midsouth', 'Northeast', 'Plains', 'South Central', 'Southeast', 'West']
avo_major_region = avo[avo['geography'].isin(major_regions)]
```

```python
Non_cities = ['Midsouth', 'Northeast', 'Plains', 'South Central', 'Southeast', 'West', 'South Carolina', 'Total U.S.', 'West Tex/New Mexico', 'Northern New England', 'California', 'Great Lakes']
avo_cities = avo[~avo['geography'].isin(Non_cities)]
avo_cities_for_graph = avo_cities.copy()
```

Question 3.

```python
avo_2017_org = avo_major_region[(avo['type'] == 'organic') & (avo['year'] == 2017)]
avo_2017_org = avo_2017_org.groupby(['geography']).sum()
avo_2017_org = avo_2017_org.sort_values('small', ascending =False)
avo_2017_org = avo_2017_org.reset_index()
avo_2017_org['geography'][0]
```

The West Region sold the most organic, small Hass avocados in 2017.

Question 4.

```python
avo['year'], avo['month'], avo['day'] = avo['date'].str.split('-', n=2).str
```

```python
avos = avo.groupby(['month']).mean()
avos = avos.sort_values('total_volume', ascending = False)
avos = avos.reset_index()
avos['month'][0]
```

The fifth month (May) has the highest average volume of avocado sales.

Question 6.

```python
cali_cities = avo[avo['geography'].isin(['Los Angeles', 'San Diego', 'Sacramento', 'San Francisco'])]
cali_cities_outside = cali_cities.copy()
```

Question 7.

```python
cali_cities = cali_cities.groupby(['geography', 'type']).mean().reset_index()
cali_for_graph = cali_cities.copy()
```

```python
cali_cities = cali_cities.pivot(index = "geography", columns = "type", values = "average_price")
```

```python
cali_cities['diff'] = cali_cities['organic'] - cali_cities['conventional']
cali_cities
```

```python
cali_cities = cali_cities.reset_index()
```

The city in CA where there is the greatest difference in price between conventional and organic avocados is San Francisco with a 72 cent difference followed by San Diego which had a 68 cent difference.

```python
cali_cities.describe()
```

```python
(ggplot(cali_cities, aes(x = "geography", y = "diff"))
+ geom_bar(stat = 'identity', fill = 'orange')
+ labs(x = 'geography', y = 'Org - Conv'))
```

Question 8.

```python
cali_for_graph = cali_for_graph.melt(id_vars=["geography", 'type'], value_vars=['small', 'large', 'xlarge'], var_name="size", value_name = 'Count')
```

```python
from plotnine import facet_wrap
cali_for_graph['size'] = pd.Categorical(cali_for_graph['size'], categories=['xlarge', 'large', 'small'])
(ggplot(cali_for_graph,
aes(
  x = "geography",
y = "Count",
  fill = "size"
))
 + theme_bw()
+ geom_col(position = "fill")
+ facet_wrap("type")
 + theme(axis_text_x=element_text(angle=45, hjust=1))
 + scale_fill_manual(values={'large':"mediumseagreen", 'small':"orangered", 'xlarge':"mediumslateblue"})
 + labs(x = 'Region of California', y = 'Proportion')
 
)
```

Question 9: Using Outside Data

```python
ca_house = pd.read_csv("housing_prices (1).csv")
```

```python
cali_avos_house = cali_cities_outside.merge(ca_house, left_on = ['geography', 'date'], right_on = ['city', 'date'], how = "inner")
```

```python
(ggplot(cali_avos_house, aes(x = "house_price", y = "average_price", color = 'city'))
+ geom_point()
+ labs(x = 'House Price', y = 'Avocado Price'))
```

There does not seem to be any real correlation between house price and avocado price.

```python
(ggplot(cali_avos_house, aes(x = "house_price", y = "total_volume", color = 'city'))
+ geom_point()
+ labs(x = 'House Price', y = 'Total Volume of Avo Sales'))
```

This visualization shows that there is a slight positive correlation between house price and volume of Avocado sales within the respective cities.

```python
(ggplot(cali_avos_house, aes(x = "total_volume", y = "average_price", color = 'city', size = "house_price"))
+ geom_point()
+ labs(x = 'Volume of Avos Sold', y = 'Avocado Price'))
```

There is a negative correlation between volume of avocados sold and avocado price.
