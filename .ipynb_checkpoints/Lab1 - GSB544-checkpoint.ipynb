The aesthetic is Income on the x-axis, life expectancy on the y-axis, the fill is world region, and the size is the population of the country.

```python
from plotnine.scales import scale_x_log10
from plotnine.scales import scale_y_log10
(ggplot(q1,
aes(
  x = "income",
  y = "life_exp",
  color = "four_regions",
    size = "population"
))
 +annotate("text", x=8000, y=65, label= "2010", size = 50, alpha = .2)
+ geom_point()
 + theme_bw()
 + scale_x_continuous(breaks = [500, 1000, 2000, 4000, 8000, 16000, 32000, 64000], trans = "log2", labels = ["500", '1000', '2000', '4000', '8000', '16K', '32K', '64K'])
 + scale_y_continuous(breaks = [20,30, 40, 50, 60, 70, 80])
  + theme(figure_size = (8, 4))
)
```
