```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
from palmerpenguins import *
```

```python
penguins = load_penguins()
```

```python
penguins['island'].unique()
```

```python
def pen(island_type):
    peng = penguins[penguins['island'] == island_type]
    plot = (ggplot(peng, aes(x="bill_length_mm", y = "bill_depth_mm")) + geom_point())
    return plot
```

```python
pen("Dream")
```
