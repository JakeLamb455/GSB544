```python
import pandas as pd
import numpy as np
from plotnine import *
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures, FunctionTransformer
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.float_format', '{:.4f}'.format)
np.set_printoptions(suppress=True, precision=4)
```

```python
completed = pd.read_csv("completed.csv")
completed
```

```python
#[County, Total Stops, Population^2]
enc = OneHotEncoder(sparse_output = False)

X_1 = completed.drop('Unemployment Rate', axis=1)
y_2 = completed[['Unemployment Rate']]


ct1 = ColumnTransformer([
    ('dummify_county', enc, ['County']),
    ('standardize', StandardScaler(), ["Year", "stops_pop"])
], remainder='drop').set_output(transform = 'pandas')

ct2 = ColumnTransformer([
   ('square stops per capita', PolynomialFeatures(degree=2), ["standardize__stops_pop"]),
], remainder='passthrough').set_output(transform = 'pandas')

knnJ2 = Pipeline(
    [('dummify',ct1), ("dummify2",ct2), ('knn', KNeighborsRegressor())]
)

param_grid = {
    'knn__n_neighbors': [3, 5, 7, 9, 11]
}

gscvJ2 = GridSearchCV(knnJ2, param_grid, cv = 5, scoring='neg_mean_squared_error')
gscvJfit2 = gscvJ2.fit(X_1,y_2)
print(ct2.fit_transform(ct1.fit_transform(X_1)).columns)
print(gscvJfit2.best_params_)
print(gscvJfit2.best_score_)
```

[County, Population, Total Stops, Average Distance*Total Stops]

```python

enc = OneHotEncoder(sparse_output = False)

X_1 = completed.drop('Unemployment Rate', axis=1)
y_2 = completed[['Unemployment Rate']]


ct1 = ColumnTransformer([
    ('dummify_county', enc, ['County']),
    ('standardize', StandardScaler(), ["Population", "Total Stops", "Average_Distance"])
], remainder='drop').set_output(transform = 'pandas')

ct2 = ColumnTransformer([
   ('interaction', PolynomialFeatures(interaction_only = True), ["standardize__Average_Distance",'standardize__Total Stops']),
], remainder='passthrough').set_output(transform = 'pandas')

knnJ2 = Pipeline(
    [('dummify',ct1), ("dummify2",ct2), ('knn', KNeighborsRegressor())]
)

param_grid = {
    'knn__n_neighbors': [3, 5, 7, 9, 11, 15, 20]
}

gscvJ2 = GridSearchCV(knnJ2, param_grid, cv = 5, scoring='neg_mean_squared_error')
gscvJfit2 = gscvJ2.fit(X_1,y_2)
print(ct2.fit_transform(ct1.fit_transform(X_1)).columns)
print(gscvJfit2.best_params_)
print(gscvJfit2.best_score_)
```
